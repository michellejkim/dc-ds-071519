```python
import pandas as pd
import matplotlib.pyplot as plt
suicide = pd.read_csv('data/master.csv')
```

```python
suicide.info()
```

```python
suicide[suicide.country.map(lambda x: x.startswith('U'))]
```

```python
suicide[~suicide.country.map(lambda x: x.startswith('U'))].country.unique()
```

```python
suicide[(suicide.country == 'United States') & (suicide.year == 2015)].shape
```

```python
suicide.groupby('age').sum()
```

```python
suicide.age = suicide.age.astype('category')
suicide.age.values
```

```python
suicide.age.cat.reorder_categories(['5-14 years','15-24 years', '25-34 years', '35-54 years','55-74 years','75+ years'], inplace = True)
```

```python
plt.bar(suicide[suicide.year == 2015].age.cat.categories,
        suicide[suicide['year'] == 2015].groupby('age').suicides_no.sum())
plt.xticks(suicide[suicide.year == 2015].age.cat.categories, rotation = 45)
plt.show()
plt.savefig('BarPlot_MJK')
```
