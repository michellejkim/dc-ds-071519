```python
import random
class  take_a_random_walk:
    def __init__(self):
        self.position = 0                   # init the position of the walker
        self.probability = random.random()  # init a probability of the next step
    def __iter__(self):
        return self                         # iterator creation
    def __next__(self):
        if self.probability < 0.0001:              # set the probability of terminating this undefined loop
            raise StopIteration             # end iteration
        else:                               # generate next step
            self.probability = random.random()  # random number generated
            if self.probability < 0.5:      # 
                self.position -= 1          # step backward
            elif self.probability > 0.5:
                self.position += 1          # step foward
            return self.position
```

```python
results = []
for num in take_a_random_walk():
    results.append(num)
```

```python
import matplotlib.pyplot as plt
plt.plot(results)
plt.title('1-D Random Walk')
plt.ylabel('Position')
plt.xlabel('Step')
plt.show()
```
